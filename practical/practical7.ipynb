```python
staff = {"Pete", "Kelly", "Jon", "Paul", "Sally", "Sue"}
directors = {"Kelly", "Rupert", "Cyril", "Jon"}
staff = staff | {"Mark", "Ringo"}
print(staff)
staff_directors = staff & directors
print(staff_directors)

staff_or_external = directors ^ staff
print(staff_or_external)
```
